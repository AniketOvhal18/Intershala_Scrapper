```python
#ipmporting all required laibraries 
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# search the job title manualy on internshala ex.Data Sicentist,Java developer and paste link here 
response = requests.get('https://internshala.com/') #paste here
response.status_code
```

```python
soup=bs(response.text)
type(soup)
```

```python
df=pd.DataFrame(columns=['Title','Company','location','ctc','skills','experinece','jd']) # creating df 
```

```python
#creating dataFrame
df 
```

```python
#scraping all links of jobs 
a_tags = soup.find_all('a', class_='job-title-href')

base_url = "https://internshala.com"
links = [base_url + a['href'] for a in a_tags]

print(links)
```

```python
#loop for scraping all the details of the job listed on job portal 
for i in links:
    url=i
    response=requests.get(url)
    response.text
    soup=bs(response.text)
    title=soup.find('h1',class_='heading_2_4 heading_title').text.strip()
    company = soup.find('a', class_='link_display_like_text').text.strip()
    location=soup.find('p',id='location_names').text.strip()
    ctc=soup.find('span',class_='desktop').text.strip()
    exp= soup.find('div',class_='item_body desktop-text').text.strip()
    jd=soup.find('div',class_='text-container').text.strip().replace('\n','')
    skills=soup.find('div',class_='round_tabs_container').text.strip().replace('\n',', ')
    df.loc[len(df)]=[title,company,location,ctc,skills,exp,jd]
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
3save scraped data to csv
file_name_with_extension= 'internshala_scraped_data' #save the file name 
df.to_csv(file_name_with_extension, index=False)
```
